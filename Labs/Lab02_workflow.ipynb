```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(2.965,3.035,30)

def method1(x):
  y = x**7 - 21*x**6 + 189*x**5 - 945*x**4 + 2835*x**3 - 5103*x**2 + 5103*x - 2187
  return y

def method2(x):
  y = (x-3)**7
  return y

def method3(x):
  y = ((((((x-21)*x+189)*x-945)*x+2835)*x-5103)*x+5103)*x-2187
  return y

y1 = method1(x)
y2 = method2(x)
y3 = method3(x)
y4 = np.polyval([1,-21,189,-945,2835,-5103,5103,-2187],x)

plt.plot(x,y1,label ='method 1')
plt.plot(x,y2,label ='method 2')
plt.plot(x,y3,label ='method 3',marker ='o')
plt.plot(x,y4,label ='method 4')
plt.legend()

print("It seems that method 2 is by far the best method")
```
